```python
import string
import glob
import nltk
import math
import os
from collections import defaultdict
import copy
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
files=glob.glob('C:\\Users\\siddh\\OneDrive - Shiv Nadar Institution of Eminence\\spring 2023\\spring 2023\\information retrieval\\assignment\\assignment_2\\corpus\\*')
for i in range(len(files)):
  name=files[i].split("\\")
  files[i]=name[-1]
# print(files)
file_id={}
for i in range(len(files)):
  file_id[i]=files[i]
print(dict(file_id))
# print(files)
```

```python
#case folding
def case_folding(lines):
  lines=lines.casefold()
  return lines
#removing punctuations
def rm_punctuation(words):
  words=words.translate(str.maketrans('','',string.punctuation))
  return words
#tkenization
def tokenization(words):
  token=nltk.word_tokenize(words)
  return token
#removing single letter tokens
def rm_single(token):
  for i in token:
    if len(i)==1:
      token.remove(i)
  return token
#removing stop words
def rm_stop_words(token):
  stop_words = set(stopwords.words('english'))
  for word in token:
    if word in stop_words:
        token.remove(word)
  return token
#stemming of tokens
def stemming(token):
  porter = PorterStemmer()
  for i in range(0,len(token)):
    token[i]=porter.stem(token[i])
  return token
#lemmatizing of tokens
def lemmatizing(token):
  for i in range(0,len(token)):
    token[i]=WordNetLemmatizer().lemmatize(token[i])
  return token
#reading file function
def read_file(file_name):
  with open('C://Users//siddh//OneDrive - Shiv Nadar Institution of Eminence//spring 2023//spring 2023//information retrieval//assignment//assignment_2//corpus//{}'.format(file_name), encoding="utf8") as f:
    lines=f.read()
    return lines
```

```python
index={}
# len_dict= {}
for i in range(0,len(files)):
  line=read_file(files[i])
  line=case_folding(line)
  text=rm_punctuation(line)
  tokens=tokenization(text)
  tokens=rm_single(tokens)
  tokens=rm_stop_words(tokens)
  tokens=stemming(tokens)
  tokens=lemmatizing(tokens)
  # len_dict[i]=len(tokens)

  for word in tokens:
    if word in index:
      if i in index[word][1]:
        index[word][1][i]=index[word][1][i]+1
      else:
        index[word][1][i]=1
        index[word][0]=index[word][0]+1
    else:
      index[word]=[]
      index[word].append(1)
      index[word].append({})
      index[word][1][i]=1
```

```python
mydict= {}
for i in range(0,len(files)):
  mydict[i]={}

for i in range(0,len(files)):
  line=read_file(files[i])
  line=case_folding(line)
  text=rm_punctuation(line)
  tokens=tokenization(text)
  tokens=rm_single(tokens)
  tokens=rm_stop_words(tokens)
  tokens=stemming(tokens)
  tokens=lemmatizing(tokens)
  for word in tokens:
    if word in mydict[i]:
      mydict[i][word]=mydict[i][word]+1
    else:
      mydict[i][word]=1

normal_dict={}
for i in range(0,len(files)):
  mylist=list(mydict[i].values())
  x=0.0
  for j in range(0,len(mylist)):
    x=x+math.pow(mylist[j],2)
  x=round(math.sqrt(x),4)
  normal_dict[i]=x
  # print(mylist)
# print(normal_dict)

#new index after normalization
new_index=copy.deepcopy(index)
for word in new_index:
  for i in new_index[word][1]:
    new_index[word][1][i]=(1+math.log10(new_index[word][1][i]))/normal_dict[i]
    new_index[word][1][i]=round(new_index[word][1][i],4)
print(new_index)
```

```python
query=case_folding(query)
query=rm_punctuation(query)
query_tokens=tokenization(query)
query_tokens=rm_single(query_tokens)
query_tokens=rm_stop_words(query_tokens)
query_tokens=stemming(query_tokens)
query_tokens=lemmatizing(query_tokens)
query_dict={}
# print(query_tokens)
for i in range(0,len(query_tokens)):
  if query_tokens[i] in query_dict:
    query_dict[query_tokens[i]][0]=query_dict[query_tokens[i]][0]+1
  else:
    query_dict[query_tokens[i]]=[1]
    #document frequency
    if query_tokens[i] not in new_index:
      query_dict[query_tokens[i]].append(0)
    else:
      query_dict[query_tokens[i]].append(index[query_tokens[i]][0])
# print(query_dict)
#tf=1+log(tf)
for word in query_dict:
  if word not in new_index:
    query_dict[word]=0
  else:
    query_dict[word][0]=1+math.log(query_dict[word][0],10)
    n=math.log((len(files)/query_dict[word][1]),10)
    query_dict[word][0]=(query_dict[word][0])*n
    query_dict[word].pop(1)
    query_dict[word]=round(query_dict[word][0],4)
# print(query_tokens)
print(query_dict)
```

```python
result_dict={}
result_list=[]
for word in query_tokens:
  if word in new_index:
    result_dict[word]=new_index[word]
    result_dict[word]=result_dict[word][1]
    for i in result_dict[word]:
      result_dict[word][i]=round(result_dict[word][i]*query_dict[word],4)
    result_list.append(result_dict[word])
print(result_dict)
result_dict={}
for i in range(0,len(result_list)):
  for j in range(0,len(files)):
    if j not in result_list[i]:
      result_list[i][j]=0
result_dict = {}
# print(result_list)
for key in set.intersection(*(set(d.keys()) for d in result_list)):
    result_dict[key] = sum(d[key] for d in result_list)
sorted_dict = dict(sorted(result_dict.items(), key=lambda item: item[1], reverse=True))
answer=list(sorted_dict.keys())[:20]
for i in range(0,len(answer)):
  print(files[answer[i]])
```
